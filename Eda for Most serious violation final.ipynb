```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings 
warnings.filterwarnings('ignore')

data = pd.read_csv("by most serious violation1.csv")
data.head()
# Assuming you have your dataset loaded as 'data'

# Calculate the sum for each row (excluding the 'Most serious violation' column)
data['Row Sum'] = data.iloc[:, 1:].sum(axis=1)

# Create a pie chart for the row sums
plt.figure(figsize=(10, 10))
plt.pie(data['Row Sum'], labels=data['Most serious violation'], autopct='%1.1f%%', startangle=140)
plt.title('Sum of Violations Over the Years (2014-2022)')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings 
warnings.filterwarnings('ignore')

data = pd.read_csv("by most serious violation1.csv")
data.head()

# Assuming you have your dataset loaded as 'data'

# Calculate the sum for each row (excluding the 'Most serious violation' column)
data['Row Sum'] = data.iloc[:, 1:].sum(axis=1)

# Sort the data by the sum of violations in descending order
data = data.sort_values(by='Row Sum', ascending=False)

# Create a bar chart for the row sums
plt.figure(figsize=(12, 6))
plt.bar(data['Most serious violation'], data['Row Sum'])
plt.xlabel('Category')
plt.ylabel('Sum of Violations (2014-2022)')
plt.title('Sum of Violations by Category Over the Years (2014-2022)')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability

plt.show()



```
