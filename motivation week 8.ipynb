```python
# Import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings 
warnings.filterwarnings('ignore')

data = {
    'Type of motivation': ['Race or ethnicity', 'Religion', 'Sexual orientation', 'Language', 'Disability', 'Sex and gender', 'Age', 'Other similar factor', 'Unknown motivation'],
    2014: [611, 429, 155, 12, 10, 22, 6, 10, 23],
    2015: [641, 469, 141, 18, 8, 12, 4, 27, 25],
    2016: [666, 460, 176, 13, 11, 24, 5, 44, 19],
    2017: [878, 842, 204, 23, 14, 32, 4, 35, 32],
    2018: [793, 657, 186, 14, 9, 54, 9, 48, 22],
    2019: [884, 613, 265, 25, 3, 56, 8, 73, 39],
    2020: [1619, 530, 258, 37, 8, 49, 5, 58, 39],
    2021: [1745, 886, 438, 33, 16, 60, 14, 101, 81],
    2022: [1950, 750, 491, 59, 15, 89, 7, 98, 117]
}

df = pd.DataFrame(data)


# Data Inspection
print("Data Information:")
print(df.info())
print("\nData Preview:")
print(df.head())

# Summary Statistics
print("\nSummary Statistics:")
print(df.describe())

# 1. Line plot to show trends over the years for different types of motivations
plt.figure(figsize=(12, 6))
for motivation in df.columns[1:]:
    plt.plot(df['Type of motivation'], df[motivation], marker='o', label=motivation)
plt.title('Trends of Hate Crimes by Type of Motivation (2014-2022)')
plt.xlabel('Type of Motivation')
plt.ylabel('Number of Hate Crimes')
plt.xticks(rotation=90)
plt.legend()
plt.grid()
plt.show()


# 2. Visualize the data using seaborn
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")
sns.lineplot(data=df.melt(id_vars='Type of motivation', var_name='Year', value_name='Hate Crimes'), x='Year', y='Hate Crimes', hue='Type of motivation')
plt.title("Hate Crimes by Type of Motivation Over the Years")
plt.xticks(rotation=45)
plt.xlabel("Year")
plt.ylabel("Number of Hate Crimes")
plt.legend(title="Motivation")
plt.show()

# 3. Compare hate crimes across motivations in a specific year (e.g., 2022)
year_2022 = df[['Type of motivation', 2022]].sort_values(by=2022, ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(data=year_2022, x=2022, y='Type of motivation', palette="viridis")
plt.title("Hate Crimes by Type of Motivation in 2022")
plt.xlabel("Number of Hate Crimes")
plt.ylabel("Motivation")
plt.show()


# 5. Identify outliers, if any, using box plots
plt.figure(figsize=(12, 6))
sns.boxplot(data=df.drop(columns='Type of motivation'), palette="Set3")
plt.title("Box Plot of Hate Crimes by Year")
plt.ylabel("Number of Hate Crimes")
plt.xticks(rotation=45)
plt.show()

# hate crimes over years
plt.figure(figsize=(12, 6))
for year in range(2014, 2023):
    sns.kdeplot(df[year], label=year)
plt.title("Distribution of Hate Crimes Over the Years")
plt.xlabel("Number of Hate Crimes")
plt.legend(title="Year")
plt.show()




#pairwise scatter plots
sns.pairplot(data=df[['Type of motivation', 2014, 2017, 2022]], hue='Type of motivation')
plt.suptitle("Pairwise Scatter Plots of Hate Crimes by Motivation")
plt.show()



```
